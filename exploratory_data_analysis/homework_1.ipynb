###### Convert Lat-Long
From **degree minute second + comapss direction** format to **signed degree** format

```python
def convert_lat_lon(x):
    direction = {'N':1, 'S':-1, 'E':1, 'W':-1}
    if str(x) == 'nan':
        return x
    else:
        x = str(x).split()
        return (int(x[0]) + int(x[1])/60.0) * direction[x[2]]

raw_data.latitude_of_capital = raw_data.latitude_of_capital.apply(convert_lat_lon)
raw_data.longitude_of_capital = raw_data.longitude_of_capital.apply(convert_lat_lon)
```

###### Cleaning date of founding/independence column

```python
import time
import datetime
def get_datetime(d):
    if (d == 'unknown') or (d ==  'none'):
        return pd.NaT
    if len(d) == 4:
        t_obj = time.strptime(d, "%Y")
        d_obj = datetime.datetime(*t_obj[:6])
        #d_obj = pd.to_datetime(d_obj)
    if len(d) >= 8:
        t_obj = time.strptime(d, "%m/%d/%Y")
        d_obj = datetime.datetime(*t_obj[:6])
        #d_obj = pd.to_datetime(d_obj)
    else:
        return pd.NaT
    return d_obj
```

```python
raw_data.date_of_founding_or_independence = raw_data.date_of_founding_or_independence.apply(get_datetime)
```

###### Convert numeric DataFrame columns of object dtype to float

```python
def string_to_float(x):
    x = x.replace('%', '')
    if x == 'unknown':
        return NaN
    else:
        return float64(x)

raw_data["literacy_rate_percentage"] = raw_data["literacy_rate_percentage"].apply(string_to_float)
raw_data["health_expenditure_or_gdp"] = raw_data["health_expenditure_or_gdp"].apply(string_to_float)
raw_data["military_expenditure_or_gdp"] = raw_data["military_expenditure_or_gdp"].apply(string_to_float)
raw_data["renewable_water_resources_cu_km"] = raw_data["renewable_water_resources_cu_km"].apply(string_to_float)
raw_data["net_migration_rate_or_1000_population"] = raw_data["net_migration_rate_or_1000_population"].apply(string_to_float)
raw_data["labor_force"] = raw_data["labor_force"].apply(string_to_float)
raw_data["internet_users_circa_2009"] = raw_data["internet_users_circa_2009"].apply(string_to_float)
raw_data["commercial_prime_lending_rate"] = raw_data["commercial_prime_lending_rate"].apply(string_to_float)
```

** Permutation test to determine the p-value:**

```python
def permutation_resampling(case, control, num_samples, statistic):
    """Returns p-value that statistic for case is different
    from statistc for control."""

    observed_diff = abs(statistic(case) - statistic(control))
    num_case = len(case)

    combined = np.concatenate([case, control])
    diffs = []
    for i in range(num_samples):
        xs = random.permutation(combined)
        diff = np.mean(xs[:num_case]) - np.mean(xs[num_case:])
        diffs.append(diff)

    pval = (np.sum(diffs > observed_diff) + np.sum(diffs < -observed_diff)) / float(num_samples)
    return pval, observed_diff, diffs
```

Permutation test of the pairs which correlate the strongest (Population and Labor Force)

```python
# data
case = np.array(raw_data[raw_data[raw_data.population.notnull()].labor_force.notnull()].population).tolist()
control = np.array(raw_data[raw_data[raw_data.population.notnull()].labor_force.notnull()].labor_force).tolist()

# find p-value by permutation resampling
pval, observed_diff, diffs = permutation_resampling(case, control, 10000, np.mean)

# make plots
pylab.title('Empirical null distribution for differences in mean')
pylab.hist(diffs, bins=100, histtype='step', normed=True)
pylab.axvline(observed_diff, c='red', label='diff')
pylab.axvline(-observed_diff, c='green', label='-diff')
pylab.xlabel('p = %.5f' % pval)
pylab.legend()
print("p-value = {}".format(pval))
```

Permutation test of the pairs which correlate the weakest (Life Expectancy and Commercial Prime Lending Rate)

```python
# data
case = np.array(raw_data[raw_data[raw_data.life_expectancy.notnull()].commercial_prime_lending_rate.notnull()].life_expectancy).tolist()
control = np.array(raw_data[raw_data[raw_data.life_expectancy.notnull()].commercial_prime_lending_rate.notnull()].commercial_prime_lending_rate).tolist()

# find p-value by permutation resampling
pval, observed_diff, diffs = permutation_resampling(case, control, 10000, np.mean)

# make plots
pylab.title('Empirical null distribution for differences in mean')
pylab.hist(diffs, bins=100, histtype='step', normed=True)
pylab.axvline(observed_diff, c='red', label='diff')
pylab.axvline(-observed_diff, c='green', label='-diff')
pylab.xlabel('p = %.5f' % pval)
pylab.legend()
print("p-value = {}".format(pval))
```

The p values show us that there is,

Produce a table showing the nearest and farthest neighbor to each nation on earth?**

**Write a short analysis of this table describing:**
- **What  kinds of similarities does your measure get right?**
- **What are the most interesting/surprising pairs to fall out of this analysis?  and**
- **Where does it goof up?**

##### Answer:

```python
sim_data = raw_data.ix[:, ["country_name", "population",
                           "life_expectancy", "gdp_ppp_in_us_dollars",
                           "area_sq_km", "land_boundaries_km",
                           "literacy_rate_percentage", "labor_force",
                           "internet_users_circa_2009"]].set_index("country_name").dropna().T
```

```python
# Normalizing features
sim_data / np.sum(sim_data.values)
sim_data.div(sim_data.sum())
sim_data_n = sim_data.div(sim_data.sum(1), axis=0)
```

```python
corr_dict = {}
for c1 in sim_data_n:
    count_dict = {}
    for c2 in sim_data_n:
        count_dict[c2] = spatial.distance.cosine(sim_data_n[c1], sim_data_n[c2])
    corr_dict[c1] = count_dict
pd.set_option('display.precision',2)
cdata = pd.DataFrame.from_dict(corr_dict)
```
